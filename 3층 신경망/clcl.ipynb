```python
#뉴런 AND 게이트
def and_gate(x1,x2):
    w1, w2, theta = 0.5, 0.5, 0.7
    temp = (x1*w1) + (x2*w2)
    if temp <= theta:
        return 0
    elif temp > theta:
        return 1
    
print(and_gate(0,0))
print(and_gate(0,1))
print(and_gate(1,0))
print(and_gate(1,1))
```

```python
# OR게이트
import numpy as np

def or_gate(x1,x2):
    x = np.array([x1,x2])
    w = np.array([0.5, 0.5])
    b = -0.2
    temp = np.sum(x*w) +b

    if temp<= 0:
        return 0
    elif temp > 0:
        return 1
def nand_gate(x1,x2):
    w1, w2, theta = 0.5, 0.5, 0.7
    temp = (x1*w1) + (x2*w2)
    if temp <= theta:
        return 1
    elif temp > theta:
        return 0
def and_gate(x1,x2):
    w1, w2, theta = 0.5, 0.5, 0.7
    temp = (x1*w1) + (x2*w2)
    if temp <= theta:
        return 0
    elif temp > theta:
        return 1
    
def xor_gate(x1,x2):
    a = or_gate(x1,x2)
    b = nand_gate(x1,x2)    
    y = and_gate(a,b)
    return y

print(xor_gate(0,0))
print(xor_gate(0,1))
print(xor_gate(1,0))
print(xor_gate(1,1))
```

```python
#Numpy 버전 1.20에서는 더이상 np.int를 지원하지 않는다
#계단함수

import numpy as np
import matplotlib.pyplot as plt

def step_func_plt(x):
    return np.array(x > 0, dtype=int)

# x축 범위 나타내기
x = np.arange(-5.0, 5.0, 0.1)
y = step_func_plt(x)
plt.plot(x, y)

# y축 범위 나타내기
plt.ylim(-0.1, 1.1)
plt.show()
```

```python
#시그모이드 함수
#exp는 자연상수 대략 2.7정도입니다.
#x에 넣으면 0.731, 2를 넣으면 0.88정도
import numpy as np
import matplotlib.pyplot as plt

def sigmod(x):
    return 1/ (1+np.exp(-x))

x = np.arange(-5.0, 5.0, 0.1)
y = sigmod(x)
plt.plot(x,y)
plt.ylim(-0.1, 1.1)
plt.show()
```

```python
# Relu함수
# 0을 넘으면 넘은 값을 그대로 출력하고, 0 이하이면 0을 출력하는함수
import numpy as np
import matplotlib.pyplot as plt

def relu(x):
    return np.maximum(0,x)

x = np.arange(-5.0, 5.0, 0.1)
y = relu(x)
plt.plot(x,y)
# 0 이하는 0만을 출력함 고로 y값에 어떠한 변화를 주더라도 달라지는 것은 없음
plt.ylim(-5,6)


plt.show()
```
