```python
my_list = [1, 2, 3, 4, 5]

def insert():
    user_input = input("Enter a value to add to the list (type 'stop' to end): ")
    return user_input

while True:
    any_value = insert()
    if any_value == "stop":
        break   
    else:
        my_list.append(any_value)
    print(my_list)
```
